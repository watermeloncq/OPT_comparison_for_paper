```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
# import seaborn as sns
# sns.set_style("darkgrid",{"font.sans-serif":['KaiTi', 'Arial']})
# plt.style.use('ggplot')
plt.rcParams['font.sans-serif']=['SimHei'] #显示中文标签
plt.rcParams['axes.unicode_minus']=False   #这两行需要手动设置
%matplotlib inline
```

```python
fig = plt.figure(figsize=(14, 2.8))
# fig.autofmt_xdate()
# plt.suptitle('OPT', fontsize=16,x=0.51,y=0.95)
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

df1 = pd.read_csv("./Stocks/data/temp/forStocks/values/Values_MV_MinRisk.csv",index_col='date',parse_dates=True)
df2 = pd.read_csv("./Stocks/data/temp/forStocks/weights/Weights_MV_MinRisk.csv",index_col='date',parse_dates=True)
df3 =pd.read_csv("./Stocks/data/temp/forStocks/costs/Costs_MV_MinRisk.csv",index_col='date',parse_dates=True)

ax1.plot(df1), ax1.set_ylabel('MV-MinRisk')
ax1.xaxis.set_major_locator(ticker.MultipleLocator(60)) # 设置日期显示间距，避免重叠
ax2.plot(df2), ax2.legend(df2.iloc[:,1:])
ax2.xaxis.set_major_locator(ticker.MultipleLocator(60))
ax3.plot(df3)
ax3.xaxis.set_major_locator(ticker.MultipleLocator(60))


fig.savefig('./img/plot_other_OPT_results_Stocks/MV-MinRisk.svg', 
                      bbox_inches='tight',
                      dpi=300,
                      format='svg')
```
